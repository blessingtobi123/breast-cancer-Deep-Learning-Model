```python
#import commonly used libraries in Python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pydicom
```

```python
# 1. Load DICOM files
file_path_1 = './1-5.dcm'    #in two-dimensional (2D) medical image (dcm)
file_path_2 = './1-9.dcm'    #in two-dimensional (2D) medical image  (dcm)
```

```python
#pydicom.dcmread to display from image 1 and 2
dicom_image_1 = pydicom.dcmread(file_path_1) 
dicom_image_2 = pydicom.dcmread(file_path_2) 
```

```python
#1a. Image 1 display
plt.subplot(1, 2, 1)
plt.imshow(dicom_image_1.pixel_array, cmap='gray')
plt.title('DICOM Image 1')
```

```python
#1b. image 2 dispaly 
plt.subplot(1, 2, 2)
plt.imshow(dicom_image_2.pixel_array, cmap='gray')
plt.title('DICOM Image 2')
#show image
plt.show()
```

```python
# 2. Fuse the images
fused_image = np.maximum(dicom_image_1.pixel_array, dicom_image_2.pixel_array)
# Display the fused image
plt.imshow(fused_image, cmap='gray')
plt.title('Fused Image')
plt.show()
```

```python
#3. Apply Fourier transform on both images and display outputs
from scipy.fft import fft2, fftshift

# Fourier transform
fft_image_1 = fftshift(fft2(dicom_image_1.pixel_array))
fft_image_2 = fftshift(fft2(dicom_image_2.pixel_array))

# Display the Fourier transformed images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(np.log(np.abs(fft_image_1) + 1), cmap='gray')
plt.title('Fourier Transform - Image 1')

plt.subplot(1, 2, 2)
plt.imshow(np.log(np.abs(fft_image_2) + 1), cmap='gray')
plt.title('Fourier Transform - Image 2')
plt.show()
```

```python
#4. Perform edge enhancement on both images and display outputs
from scipy.ndimage import gaussian_filter, sobel

# Edge enhancement
edge_enhanced_image_1 = sobel(dicom_image_1.pixel_array)
edge_enhanced_image_2 = sobel(dicom_image_2.pixel_array)

# Display the edge-enhanced images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(edge_enhanced_image_1, cmap='gray')
plt.title('Edge Enhancement - Image 1')

plt.subplot(1, 2, 2)
plt.imshow(edge_enhanced_image_2, cmap='gray')
plt.title('Edge Enhancement - Image 2')

plt.show()
```

```python
#Rotate the 2 images at angle 45 and display your outputs
from scipy.ndimage import rotate

# Rotate the images at angle 45
rotated_image_1 = rotate(dicom_image_1.pixel_array, angle=45)
rotated_image_2 = rotate(dicom_image_2.pixel_array, angle=45)

# Display the rotated images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(rotated_image_1, cmap='gray')
plt.title('Rotated Image - Image 1')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image_2, cmap='gray')
plt.title('Rotated Image - Image 2')

plt.show()
```

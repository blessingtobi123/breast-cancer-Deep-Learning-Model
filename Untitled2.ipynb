```python
import os
import cv2
import matplotlib.pyplot as plt
import pydicom

# Path to the folder containing breast cancer images in DICOM format
data_folder = r'C:\Users\USER\Desktop\codes\image\non'

# Create a new folder for preprocessed images
output_folder = r'C:\Users\USER\Desktop\codes\image\non_preprocessed'
os.makedirs(output_folder, exist_ok=True)

def preprocess_dicom(dicom_path):
    # Read the DICOM image
    dicom_image = pydicom.dcmread(dicom_path)

    # Ensure the image is in grayscale (single channel)
    if dicom_image.pixel_array.ndim > 2:
        # Convert to grayscale
        resized_image = cv2.cvtColor(dicom_image.pixel_array, cv2.COLOR_BGR2GRAY)
    else:
        resized_image = dicom_image.pixel_array

    # Resize the image to a standard size (e.g., 256x256)
    resized_image = cv2.resize(resized_image, (256, 256))

    # Ensure the image is in 8-bit unsigned integer format
    resized_image = cv2.convertScaleAbs(resized_image)

    # Normalize pixel values to the range [0, 1]
    normalized_image = resized_image / 255.0

    # Apply histogram equalization for contrast enhancement
    equalized_image = cv2.equalizeHist(resized_image)

    # Apply Gaussian blur for noise reduction
    denoised_image = cv2.GaussianBlur(resized_image, (5, 5), 0)

    # Save the preprocessed DICOM image to the output folder
    output_path = os.path.join(output_folder, os.path.basename(dicom_path) + '.png')
    cv2.imwrite(output_path, denoised_image)

    return denoised_image, output_path

# Process each DICOM file in the data folder
for filename in os.listdir(data_folder):
    if filename.endswith('.dcm'):
        dicom_path = os.path.join(data_folder, filename)
        preprocessed_image, output_path = preprocess_dicom(dicom_path)

        # Display original and preprocessed DICOM images (for visualization purposes)
        plt.figure(figsize=(8, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(pydicom.dcmread(dicom_path).pixel_array, cmap='gray')
        plt.title('Original DICOM Image')

        plt.subplot(1, 2, 2)
        plt.imshow(preprocessed_image, cmap='gray')
        plt.title('Preprocessed Image')

        plt.show()

        # Print the output folder for the processed image
        print(f"Processed image saved to: {output_path}")
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pydicom

# Path to the folder containing breast cancer images in DICOM format
data_folder = r'C:\Users\USER\Desktop\codes\image\non'

# Create a new folder for preprocessed images
output_folder = r'C:\Users\USER\Desktop\codes\image\non_preprocessed'
os.makedirs(output_folder, exist_ok=True)

def preprocess_dicom(dicom_path):
    # Read the DICOM image
    dicom_image = pydicom.dcmread(dicom_path)

    # Ensure the image is in grayscale (single channel)
    if dicom_image.pixel_array.ndim > 2:
        # Convert to grayscale
        resized_image = cv2.cvtColor(dicom_image.pixel_array, cv2.COLOR_BGR2GRAY)
    else:
        resized_image = dicom_image.pixel_array

    # Resize the image to a standard size (e.g., 256x256)
    resized_image = cv2.resize(resized_image, (256, 256))

    # Ensure the image is in 8-bit unsigned integer format
    resized_image = cv2.convertScaleAbs(resized_image)

    # Normalize pixel values to the range [0, 1]
    normalized_image = resized_image / 255.0

    # Apply histogram equalization for contrast enhancement
    equalized_image = cv2.equalizeHist(resized_image)

    # Apply Contrast Limited Adaptive Histogram Equalization (CLAHE)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    clahed_image = clahe.apply(resized_image)

    # Apply Retinex (Simple White Balancing)
    retinex = cv2.ximgproc.createSimpleWB()
    retinex_image = retinex.apply(resized_image)

    # Apply Unsharp Masking
    blurred = cv2.GaussianBlur(resized_image, (0, 0), 5)
    unsharp_mask = cv2.addWeighted(resized_image, 1.5, blurred, -0.5, 0)

    # Apply Gamma Correction
    gamma = 1.5
    gamma_corrected = np.power(resized_image, gamma)

    # Save the preprocessed DICOM images to the output folder
    output_path = os.path.join(output_folder, os.path.basename(dicom_path) + '_preprocessed.png')
    cv2.imwrite(output_path, gamma_corrected)

    return gamma_corrected, output_path

# Process each DICOM file in the data folder
for filename in os.listdir(data_folder):
    if filename.endswith('.dcm'):
        dicom_path = os.path.join(data_folder, filename)
        preprocessed_image, output_path = preprocess_dicom(dicom_path)

        # Display original and preprocessed DICOM images (for visualization purposes)
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 3, 1)
        plt.imshow(pydicom.dcmread(dicom_path).pixel_array, cmap='gray')
        plt.title('Original DICOM Image')

        plt.subplot(1, 3, 2)
        plt.imshow(preprocessed_image, cmap='gray')
        plt.title('Preprocessed Image')

        plt.subplot(1, 3, 3)
        plt.imshow(gamma_corrected, cmap='gray')
        plt.title('Gamma Corrected Image')

        plt.show()

        # Print the output folder for the processed image
        print(f"Processed image saved to: {output_path}")
```
